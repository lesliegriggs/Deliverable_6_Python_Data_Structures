```python
import pandas as pd
```

```python
a = int(input("Please enter an integer value greater than 0: "))

if a <= 0:
  print('That integer is less than 0. Try again')
else:
  b = int((2*a))
  c = int(a - 2)
  evenList = []
  oddList = []
  fibList = [0, 1]

  for i in range(0, b):
    if i %2 == 0:
      evenList.append(i)
    else:
      oddList.append(i)

  for i in range(0, c):
    d = fibList[-1] + fibList[-2]
    fibList.append(d)

  dataFrame = pd.DataFrame({'Even': evenList, 'Odd': oddList, 'Fibonacci': fibList})

dataFrame

    
  
```
